#### Attrition of Employee from Research and development environment

```python
Attrition_age = df.Age.between(25,35)
Attrition_Department = df.Department == "Research & Development"
Monthly_income  = df.MonthlyIncome.between(3000, 5000)

employee_attrition = df.loc[Attrition_age & Attrition_Department & Monthly_income, ['MaritalStatus','Age']]
employee_attrition

```
