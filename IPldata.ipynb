```python
import os
import pandas as pd
from matplotlib import pyplot as plottingData
```

```python

# Set the directory where the CSV files are stored
csv_dir = r'E:\Kaggle DB\Extracted\IPL'

# Get a list of all CSV files in the directory
csv_files = [os.path.join(csv_dir, file) for file in os.listdir(csv_dir) if file.endswith('.csv')]

# Create an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Iterate over each CSV file and append its contents to the merged DataFrame
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    merged_df = pd.concat([merged_df, df])

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('all_Data_in_cricket.csv', index=False)

merged_df.to_csv('E:/Kaggle DB/Extracted/IPL/all_Data_in_cricket.csv', index=False)

```
